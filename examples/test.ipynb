```python
first=metric('first',w_size=100)
second=metric('second',w_size=100)
third=metric('third',w_size=5)

new_cm=traintorch(n_custom_plots=3,main_grid_hspace=.1,
              custom_window=[20,5,10],window=5,figsize=(15,15))
```

```python
elapsed=[]

test=[]
for i in range(0,5000,1):

    actual=np.random.choice([0, 1], size=(20,), p=[1./3, 2./3])
    predicted=np.random.choice([0, 1], size=(20,))
    
    

    t0 = time.time()

    first.update(x=i,b=i*100,g=i**2)
    second.update(train=1/(i+1),test=1/(i**2+1))
    
    if((i+1)%20==0 and i>0):
        third.update(result=np.sin(i%3))
        new_cm.create(custom_metrics=[first,second,third])
        t1 = time.time()
        elapsed.append(t1-t0)        
```
